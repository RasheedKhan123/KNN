```python
def standardize(df):
    from sklearn.preprocessing import StandardScaler
    s = StandardScaler()
    return s.fit_transform(df)
```

```python
df_scaled = pd.DataFrame(standardize(df))
```

```python
df.shape
df.head(5)
```

```python
from sklearn.decomposition import PCA
pca = PCA()
df_polied = pd.DataFrame(pca.fit_transform(df_scaled))
evr = pca.explained_variance_ratio_
```

```python
import matplotlib.pyplot as plt
plt.bar(np.arange(1, len(evr) + 1), evr)
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance Ratio by Principal Component')
plt.show()
```

```python
plt.plot(np.arange(1, len(evr) + 1), np.cumsum(evr), marker='o')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.title('Cumulative Explained Variance Ratio by Principal Components')
plt.grid(True)
plt.show()
```

```python
np.argmax(np.cumsum(evr) >= 0.8) + 1
```

(and any other PCA feature) is a linear combination of the old 23 continuous features.
Find out the exact linear combination. What old features have the largest coefficients in
the linear combination? These features are the dominant old features for the first PCA
dimension.

i.Make a new dataframe, pca df, of the food items with dP CA continuous features

```python
pca_df = df_polied.iloc[:,:np.argmax(np.cumsum(evr) >= 0.8) + 1]
```

```python
pca_df.shape
```

```python
# Reference i seached in google.

from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=10)
    kmeans.fit(pca_df)
    labels = kmeans.labels_
    avg = silhouette_score(pca_df, labels)
    scores.append(avg)

# Plot the silhouette scores
plt.plot(range(2, 11), scores, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Analysis for Optimal k')
plt.show()
```

```python
# Reference i seached in google.

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Run K-means for different values of k
inertias = []
for k in range(1, 20):
    kmeans = KMeans(n_clusters=k, random_state=10)
    kmeans.fit(pca_df)
    inertias.append(kmeans.inertia_)

# Plot the elbow method graph
plt.plot(range(1, 20), inertias, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=10)
pca_df['Cluster'] = kmeans.fit_predict(pca_df.iloc[:, :-2])
```

```python
pca_df
```

```python
pca_df['Cluster'].value_counts()
```

```python
largest_clusters = pca_df['Cluster'].value_counts().nlargest(2).index
largest_clusters
```

```python
df = pd.read_csv("food-nutrients (1).csv")
pca_df['FoodGroup'] = df['FoodGroup']
```

```python
dominant_food_groups = []
for cluster_label in largest_clusters:
    cluster_df = pca_df[pca_df['Cluster'] == cluster_label]
    food_group_counts = cluster_df['FoodGroup'].value_counts()
    
    print(f"\nCluster {cluster_label} - Food Group Counts:")
    print(food_group_counts)
    
    # Identify the dominant food group
    dominant_food_group = food_group_counts.idxmax()
    dominant_food_groups.append(dominant_food_group)

# Display the dominant food groups and suggest names for the clusters
for i, cluster_label in enumerate(largest_clusters):
    print(f"\nCluster {cluster_label} - Dominant Food Group: {dominant_food_groups[i]}")
    print(f"Suggested Name: {dominant_food_groups[i]} Cluster")
```
